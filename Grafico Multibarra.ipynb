```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv ("C:/Users/USUARIO/Downloads/comparasion_algoritmos.csv")
df
```

```python
ax = df.plot(
    kind = "bar",
    x = "dataset",
    y = ["algoritmo1" , "algoritmo2"],
    figsize = (12, 3)
)

ax.grid(visible=True)
ax.set_xlabel("Dataset")
ax.set_ylabel("Rendimiento")
ax.set_title("Evaluacion de Algoritmo 1 vs Algoritmo 2")
ax.set_axisbelow(True)
ax.legend(
    labels = ["Algoritmo 1" , "Algoritmo 2"]

)
```
